```python
import pandas as pd
import numpy as np
import sklearn.datasets as sk
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data=pd.read_csv('spam.csv')
```

```python
# pd.set_option('future.no_silent_downcasting', True)
```

```python
# data=data.replace({'spam': 0, 'ham': 1})
```

```python
y=data['Category']
x=data['Message']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(y_train)
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)

y_train =np.array(y_train).astype('object')
y_test = np.array(y_test).astype('object')
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train_features,y_train)
```

```python
x_train_prediction=model.predict(x_train_features)
training_data_accuracy=accuracy_score(y_train,x_train_prediction)
```

```python
print(training_data_accuracy)
```

```python
model.fit(x_test_features,y_test)
```

```python
x_test_prediction=model.predict(x_test_features)
testing_data_accuracy=accuracy_score(y_test,x_test_prediction)
```

```python
print(testing_data_accuracy)
```

```python
print(x_train_features.shape,y_train.shape)
```

```python
raw_input = ["WINNER!! As a valued network customer you have been selected to receivea £900 prize reward! To claim..."]
input_features = feature_extraction.transform(raw_input)
prediction = model.predict(input_features)
print(prediction)
```
